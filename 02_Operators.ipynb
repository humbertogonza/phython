```python
### Operadores Aridmeticos ###

# Operaciones con enteros
print(3 + 4)
print(3 - 4)
print(3 * 4)
print(3 / 4)
print(10 % 3)
print(10 // 3)
print(2 ** 3)
print(2 ** 3 + 3 - 7 / 1 // 4)

# Operaciones con cadenas de texto
print("Hola " + "Python " + "¿Que tal?")
print("Hola " + str(5))

# Operaciones mixtas
print("Hola " * 5)
print("Hola " * (2 ** 3))

my_float = 2.5 * 2
print("Hola " * int(my_float))

### Operadores Com´parativos ###

# Operadores con enteros
print(3 > 4)
print(3 < 4)
print(3 >= 4)
print(4 <= 4)
print(3 == 4)
print(3 != 4)

# Operaciones con cadenas de texto
print("Hola" > "Python")
print("Hola" < "Python")
print("aaaa" >= "abaa") # Ordenacion alfabetica por ASCII
print(len("aaaa") >= len("abaa")) # Cuenta caracteres
print("Hola" <= "Python")
print("Hola" == "Hola")
print("Hola" != "Python")




###Operadores logicos ###

# basada en la algebra de boole https://ea.wikipedia.org/wiki/%c3&811gebra de boole
print(3 > 4 and "Hola" > "python")
print(3 > 4 or "Hola" > "python")
print(3 > 4 and "Hola" > "python")
print(3 > 4 or "Hola" > "python")
print(3 > 4 or ("Hola" > "python" and 4 == 4))
print(not(3 > 4))
```
